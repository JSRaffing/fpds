```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
```

```python
#fpds - 2023 data api call
data_2023 = pd.read_csv("https://www.fpds.gov/ezsearch/fpdsportal?s=FPDS&indexName=awardfull&templateName=CSV&q=++CONTRACT_FISCAL_YEAR%3A%222023%22+PRODUCT_OR_SERVICE_CODE%3A%22D3*%22",encoding="latin-1")
```

```python
#fpds - 2024 data api call
data_2024 = pd.read_csv("https://www.fpds.gov/ezsearch/fpdsportal?s=FPDS&indexName=awardfull&templateName=CSV&q=++CONTRACT_FISCAL_YEAR%3A%222024%22+PRODUCT_OR_SERVICE_CODE%3A%22D3*%22",encoding="latin-1")
```

```python
#convert type of action obligation to int but clean it up first
data_2023['Action Obligation ($)'] = data_2023['Action Obligation ($)'].str.replace("$", "")
data_2023['Action Obligation ($)'] = data_2023['Action Obligation ($)'].str.replace(",", "")
data_2023['Action Obligation ($)'] = data_2023['Action Obligation ($)'].astype(float)
```

```python
#2023 contract sums
grouped_data_2023 = data_2023.groupby('Contracting Agency')['Action Obligation ($)'].sum()
grouped_data_2023.sort_values(ascending=False).head(10).plot(kind='bar')
plt.title('Total Amount by Contracting Agency')
plt.xlabel('Contracting Agency')
plt.ylabel('Total Amount')
plt.show()
```

```python
#2024 contract sums
#convert type of action obligation to int but clean it up first
data_2024['Action Obligation ($)'] = data_2024['Action Obligation ($)'].str.replace("$", "")
data_2024['Action Obligation ($)'] = data_2024['Action Obligation ($)'].str.replace(",", "")
data_2024['Action Obligation ($)'] = data_2024['Action Obligation ($)'].astype(float)
grouped_data_2024 = data_2024.groupby('Contracting Agency')['Action Obligation ($)'].sum()
grouped_data_2024.sort_values(ascending=False).head(10).plot(kind='bar')
plt.title('Total Amount by Contracting Agency')
plt.xlabel('Contracting Agency')
plt.ylabel('Total Amount')
plt.show()
```

```python
#which month has the most contracts signed in 2023
data_2023['Date Signed'] = pd.to_datetime(data_2023['Date Signed']) #converting column to date time
data_2023['Month_Signed'] = data_2023['Date Signed'].dt.strftime('%B') #adding column that has month name
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

data_2023['Month_Signed'] = pd.Categorical(data_2023['Month_Signed'], categories=month_order, ordered=True)
date_counts = data_2023.groupby('Month_Signed').size()

# Plot the scatter plot
plt.figure(figsize=(12, 10))
plt.scatter(date_counts.index, date_counts.values)
plt.title('Number of Rows by Year')
plt.xlabel('Year')
plt.ylabel('Number of Rows')
plt.grid(True)
plt.show()
```

```python
#which month has the most contracts signed in 2024
data_2024['Date Signed'] = pd.to_datetime(data_2024['Date Signed']) #converting column to date time
data_2024['Month_Signed'] = data_2024['Date Signed'].dt.strftime('%B') #adding column that has month name
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

data_2024['Month_Signed'] = pd.Categorical(data_2024['Month_Signed'], categories=month_order, ordered=True)
date_counts = data_2024.groupby('Month_Signed').size()

# Plot the scatter plot
plt.figure(figsize=(12, 10))
plt.scatter(date_counts.index, date_counts.values)
plt.title('Number of Rows by Year')
plt.xlabel('Year')
plt.ylabel('Number of Rows')
plt.grid(True)
plt.show()
```

```python
#adding columns to each that represent year
data_2023['Fiscal Year'] = '2023'
data_2024['Fiscal Year'] = '2024'
```

```python
#create database
conn = sqlite3.connect("fpds")
c = conn.cursor()
```

```python
data = pd.concat([data_2023, data_2024], ignore_index=True)
```

```python
#create table
```

```python
data.to_sql('fpds', conn, if_exists='replace', index=False)
```
